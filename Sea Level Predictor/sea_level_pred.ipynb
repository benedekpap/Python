# Sea Level Predictor

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from scipy import stats
```

```python
df = pd.read_csv('epa-sea-level.csv')
df
```

```python
fig, ax = plt.subplots()
df.plot(kind ='scatter', x='Year', y='CSIRO Adjusted Sea Level', s=10, ax=ax)
ax.set_xticks(list(range(min(df['Year']), max(df['Year']), 25)))
plt.show()
```

```python
lin_regr_all = stats.linregress(df.Year, df['CSIRO Adjusted Sea Level'])
regr_line_all_x = np.linspace(np.min(df.Year), 2050)
regr_line_all_y = regr_line_all_x * lin_regr_all[0] + lin_regr_all[1]

lin_regr_2000 = stats.linregress(df.Year[120:], df['CSIRO Adjusted Sea Level'][120:])
regr_line_2000_x = np.linspace(2000, 2050)
regr_line_2000_y = regr_line_2000_x * lin_regr_2000[0] + lin_regr_2000[1]
```

```python
fig, ax = plt.subplots()
df.plot(kind ='scatter', x='Year', y='CSIRO Adjusted Sea Level', s=10, ax=ax)
ax.set_xticks(list(range(min(df['Year']), 2050, 20)))
ax.plot(regr_line_all_x, regr_line_all_y)
ax.plot(regr_line_2000_x, regr_line_2000_y)
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

def draw_plot():
    # Read data from file
  df = pd.read_csv('epa-sea-level.csv')
    

    # Create scatter plot
  fig, ax = plt.subplots()
  df.plot(kind ='scatter', x='Year', y='CSIRO Adjusted Sea Level',\
            s=10, ax=ax)
  ax.set_xticks(list(range(min(df['Year']), 2050, 20)))

    # Create first line of best fit
  lin_regr_all = stats.linregress(df.Year, df['CSIRO Adjusted Sea Level'])
  regr_line_all_x = np.linspace(np.min(df.Year), 2050)
  regr_line_all_y = regr_line_all_x * lin_regr_all[0] + lin_regr_all[1]
  
  ax.plot(regr_line_all_x, regr_line_all_y)

    # Create second line of best fit
  lin_regr_2000 = stats.linregress(df.Year[120:],\
                                   df['CSIRO Adjusted Sea Level'][120:])
  regr_line_2000_x = np.linspace(2000, 2050)
  regr_line_2000_y = regr_line_2000_x * lin_regr_2000[0] + lin_regr_2000[1]

  ax.plot(regr_line_2000_x, regr_line_2000_y)


    # Add labels and title
  ax.set_ylabel('Sea Level (inches)')
  ax.set_title('Rise in Sea Level')

    
    # Save plot and return data for testing (DO NOT MODIFY)
  plt.savefig('sea_level_plot.png')
  return plt.gca()
```

```python
draw_plot()
```
